```python
import pandas as pd
```

```python
hrt = pd.read_csv('../gen/analisis/hrt.csv', index_col=0)
hrt_percent = hrt.apply(lambda x: x/hrt.loc['nbr'], axis=1)
hrt_percent.to_csv('../gen/analisis/hrt_percent.csv')

hrt_percent = hrt_percent.iloc[:-1]
del hrt_percent['scale']
hrt_percent.plot(figsize=(20, 15)).get_figure().savefig('../gen/analisis/hrt.jpeg')
```

```python
modis_hrt = pd.read_csv('../gen/analisis/modis_hrt.csv', index_col=0)
modis_percent = modis_hrt.apply(lambda x: x/hrt.loc['nbr'], axis=1)
modis_percent.to_csv('../gen/analisis/modis_percent.csv')

modis_percent = modis_percent.iloc[:-1]
del modis_percent['scale']
modis_percent.plot(figsize=(20, 15)).get_figure().savefig('../gen/analisis/modis_hrt.jpeg')
```
