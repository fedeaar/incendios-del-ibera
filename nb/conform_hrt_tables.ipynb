```python
import pandas as pd
```

```python
nbr = pd.read_json('../gen/gee/sentinel-2 nbr/nbr_areas.json')
hrt_02 = pd.read_json('../gen/gee/sentinel-2 hrt/hrt_02-2022.json')
hrt_03 = pd.read_json('../gen/gee/sentinel-2 hrt/hrt_03-2022.json')
hrt_04 = pd.read_json('../gen/gee/sentinel-2 hrt/hrt_04-2022.json')
hrt_05 = pd.read_json('../gen/gee/sentinel-2 hrt/hrt_05-2022.json')
hrt_06 = pd.read_json('../gen/gee/sentinel-2 hrt/hrt_06-2022.json')
hrt = pd.concat([
    hrt_02.iloc[0].rename('02-27'), 
    hrt_03.iloc[0].rename('03-31'), 
    hrt_04.iloc[0].rename('04-30'), 
    hrt_05.iloc[0].rename('05-31'), 
    hrt_06.iloc[0].rename('06-30'), 
    nbr.iloc[0].rename('nbr')
], axis=1).T
hrt.to_csv('../gen/analisis/hrt.csv')
```
